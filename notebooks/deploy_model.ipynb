```python
import mlflow
from mlflow.tracking import MlflowClient
import os
import joblib, json
```

```python
mlflow_tracking_path = os.path.abspath(os.path.join(os.getcwd(),os.path.pardir,'mlruns'))
mlflow.set_tracking_uri("file:\\"+mlflow_tracking_path)
```

```python

MODEL_NAME = "COPD_Classifier"
OUTPUT_PATH =os.path.abspath(os.path.join(os.getcwd(),os.path.pardir,"models"))
client = MlflowClient()
prod_versions = client.get_latest_versions(MODEL_NAME, stages=["Production"])

if not prod_versions:
    raise ValueError("No Production model found in MLflow registry.")

prod_model = prod_versions[0]
model_uri = f"models:/{MODEL_NAME}/Production"
version = prod_model.version
run_id = prod_model.run_id
metadata = {
    "Model_name": MODEL_NAME,
    "version": version,
    "run_id": run_id    
}
# Load model and save to joblib
model = mlflow.sklearn.load_model(model_uri)
joblib.dump(model, os.path.join(OUTPUT_PATH,MODEL_NAME))
with open(os.path.join(OUTPUT_PATH,"model_metadata.json"),"w") as f:
    json.dump(metadata,f)
```
