```python
import requests
```

```python
API_URL = 'https://a83rxleacg.execute-api.us-east-1.amazonaws.com/datadoctor-chat'
```

```python
def chat_fn(message):
    r = requests.post(API_URL, json={"query": message})
    answer = r.json().get("answer", "Error")
    return answer
```

```python
message = "What kind of surgery was Jane Smith submitted to?"
ans=chat_fn(message)
print(ans)
```
